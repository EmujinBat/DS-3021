## Question 1

### Question: Among the people with master's degree, are there a difference in the income between genders?

### Pseudocode: 

- Isolate people with master's degree
- Group by gender
- Group by gender and count the number of people in total
- Group by gender and count the number of people who earns >50K
- Find the proportions by dividing the number of people who earns more than 50k by the total count


Preparing the data frame first: 

```python
# Renaming columns
adult_data.rename(columns={' Male': 'Gender', ' <=50K': 'Income Level', ' Never-married': 'Marriage Status', ' Bachelors': 'Education Level', ' Adm-clerical': 'Occupation'}, inplace=True)
adult_data.head(3)
```

```python
# Changing data types of relevant columns
adult_data = adult_data.astype({'Gender': 'category', 'Income Level': str, 'Marriage Status': 'category', 'Education Level': str})

print(adult_data.info())
```

### Single Line

Creating a new data frame with only people with master's degree

```python
adult_masters = adult_data.query("`Education Level` == ' Masters'")
print(len(adult_masters))
adult_masters.head(5)
```

```python
# Counting how many people are in each gender regardless of income
# .size() counts the number of rows when called on a GroupBy object 
total_count_bygender = adult_masters.groupby('Gender').size()
```

```python
# Counting the number of people with income above 50k in each gender

high_income_count = adult_masters[adult_masters['Income Level'] == ' >50K'].groupby('Gender').size()
```

```python
proportion = high_income_count / total_count_bygender
print(proportion)
```

### Function:

```python
def income_gap(x):
    # x1 is the masters_data where we isolated people with masters degree
    x1 = x.query("`Education Level` == ' Masters'")
    total1 = x1.groupby('Gender').size() # counting by gender
    total2 = x1[x1['Income Level'] == ' >50K'].groupby('Gender').size() # counting by gender and high income
    p = (total2 / total1)
    
    return p


    
```

```python
print(income_gap(adult_data))
```

Answer to Q1: The proportion suggests that about 65.7% of Males make more than $50K a year while only 33.3% of Females make more than $50K a year even though they have the same education level (masters degree). Among the people with master's degree, there is a significant difference in the income between genders.

## Question 2

### Question: Are there differences in the number of people who make over $50k by Marriage Status?

target variable - ratio of people who make over $50K

### Pseudocode: 

- Group by marriage status
- Count total high income people
- Calculate % high income people in each groups

### Single Line: 

```python
# Slicing the main dataframe to make it simpler
adult1 = adult_data[['Marriage Status', 'Income Level']]
```

```python
adult2 = adult1.groupby('Marriage Status')

# Initializing a new DataFrame
df = pd.DataFrame()
df = df.assign(marriagestat_total=adult2.size(), # Total number of people in each groups
               
               # Number of people who earn >50K in each group
            income_count = (adult1['Income Level'] == ' >50K').groupby(adult1['Marriage Status']).sum(),
            
               # Calculating the % in each group
            percentage = lambda df: (df['income_count'] / df['marriagestat_total']))
df = df.sort_values(by='percentage')
print(df)
```

### Function: 

```python
def income_gap(data):
    adult1 = data[['Marriage Status', 'Income Level']]
    adult2 = adult1.groupby('Marriage Status')
    df = pd.DataFrame()
    df = df.assign(marriagestat_total=adult2.size(),
                   income_count = (adult1['Income Level'] == ' >50K').groupby(adult1['Marriage Status']).sum(),
                   percentage = lambda df: (df['income_count'] / df['marriagestat_total']))
    df = df.sort_values(by='percentage')
    
    return df
```

```python
income_gap(adult_data)
```

Answer to Q2: The dataframe suggests that on average, people who are married to a member of the Armed Forces or a Civilian are likely to earn >50K than people who have different marriage statuses. According to the data, people who were never-married are the least likely to earn over 50K.

## Question 3
### Question: Do pull requests vary by quarters within a year? 

### Sudo code: 
- Change data type to categorical
- Group by quarter
- Create a new dataframe
- Compute the sum of pull requests in each group

### Single Line: 

```python
# Convert the data type first so we can group 
github_stats['quarter'] = github_stats['quarter'].astype("category")

github_df = pd.DataFrame()

# Group by the quarters and compute sum of the count column
github_df = github_df.assign(
    total_pull = github_stats.groupby('quarter')['count'].sum()
)

github_df = github_df.sort_values(by='total_pull')

print(github_df)
```

### Function

```python
def func4(data3): 
    data3['quarter'] = data3['quarter'].astype("category")
    github_df = pd.DataFrame()
    github_df = github_df.assign(total_pull = data3.groupby('quarter')['count'].sum())
    github_df = github_df.sort_values(by='total_pull')
    
    return github_df

print(func4(github_stats))



```

Answer to Q3: The data tells us that, the sum of the pull requests made in each quarters do not have a significant difference.

### Function: 

```python
def func5(data4): 
    data4['name'] = data4['name'].astype("category")
    github_df2 = pd.DataFrame()
    github_df2 = github_df2.assign(pull_lang = data4.groupby('name')['count'].sum())
    github_df2 = github_df2.sort_values(by='pull_lang', ascending=False)
    
    return github_df2

print(func5(github_stats))
```
